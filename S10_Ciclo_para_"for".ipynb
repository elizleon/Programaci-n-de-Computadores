El ciclo ``for`` (para). Esta estructura de control tiene dos propositos primordiales que no siempre son soportados por todo lenguaje de programación:

 
*  Para escribir un ciclo mientras ``while`` de forma **compacta**

*  Para **iterar** sobre los elementos de una colección de elementos.

Se utiliza cuando se conocen los valores inicial y final de la variable que controla la condición de parada.


## Ciclo ``for`` como versión compacta del ciclo ``while``

En muchos lenguajes (lamentablemente en Python no es de esta manera), el ciclo para (``for``) se usa para escribir de manera compacta un ciclo mientras (``while``). 

```
<bloque_prev>
<inicia>
while(<cond>){
   <bloque>
   <actualiza>
}
<bloque_sigui>
```

Un esquema textual que en C++ o Java representa dicho ciclo mientras (while) usando un ciclo para (for) es el que se da en el siguientefragmento de código.
 ```
<bloque_prev>
for(<inicia>; <cond>; <actualiza>){
    <bloque>
}
<bloque_sigui>
```

## Ciclo ``for`` como iterador de colecciones

Un ciclo para (``for``) puede ser usado (y en Python es su único uso) para obtener uno a uno los elementos de una colección de elementos y poder realizar con cada uno de ellos el mismo bloque de operaciones.
Un esquema textual que en Python representa dicho ciclo para (``for``) es el que se da en el siguiente fragmento de código.
```
<bloque_prev>
for <elemento> in <coleccion>:
  <bloque>
<bloque_sigui>
```
en donde:

El fragmento ``<bloque_prev>`` es el bloque de instrucciones previas que han sido ejecutadas antes del ciclo.

El fragmento ``<elemento>`` es la variable que se usa para ir recorriendo (iterando) sobre los elementos de la colección (tomará como valor cada uno de ellos en cada iteración).

El fragmento ``<coleccion>`` es la colección de elementos que será recorrida (iterada) con el ciclo.

El fragmento ``<bloque>`` es el bloque de instrucciones principal del ciclo que se ejecuta con cada uno de los elementos de la colecci ́on.

El fragmento ``<bloque_sigui>`` es el bloque de instrucciones que se ejecutan después de terminar de ejecutar el ciclo.

**El ciclo para (``for``) se puede leer como: “Para cada elemento en la colección realice las instrucciones en el ``<bloque>``”.**

###Ejemplo

````python
frutas = ["Tomate de árbol", "Maracuyá ", "Guayaba"]

for f in frutas: #para cada elemento f en la lista de frutas
  print(f)
````

Una lista es una colección de elementos que se puede definir en python. En el ejemplo, **frutas** es una lista de cadenas. Se usan las llaves [] para definir los elementos, y los elementos al ser cadenas van entre comillas.

### Break en el ciclo for
Si se requiere detener el ciclo antes de iterar sobre toda la colección de elementos se debe usar la sentencia ``break``.

####Ejemplo
Recorrer una lista hasta encontrar un elemento especial.
Dada la lista de frutas ``["Pera", "Maracuy ́a ", "Guayaba", "Lulo", "Granadilla"]`` imprimir sus elementos hasta encontrar la fruta "Guayaba".


````python
frutas = ["Tomate de árbol", "Maracuyá ", "Guayaba", "Lulo", "Granadilla"]

for f in frutas:
  print(f)
  if f == "Guayaba":
    break
````

## Colección Rango ``range``

Existen varias colecciones que se pueden iterar en Python, una de ellas es la colección Rango (``range``). 

Una colección Rango (``range``) es una secuencia de números en un intervalo (rango) semi-abierto desde ``start`` hasta ``stop`` con un incremento de ``step``, definido por:

**un valor inicial** ``start`` (lado cerrado del intervalo) que permite fijar el valor de inicio del rango.

**un valor final** ``stop``(que no se incluye en el rango, es decir el lado abierto del intervalo) define cuando termina el rango; y 

**un valor de incremento/decremento** ``step`` usado a partir del valor inicial para definir los valores del rango, es decir el incremento, por defecto se asume 1 (ejemplo: 0,1,2,3,4...). También se puede utilizar un valor negativo para ajustar la dirección del rango.

Si no se da el valor de inicio, éste se fija en cero (0) y si no se da valor de incremento/decremento, éste se fija en uno (1).

Maneja una sintaxis de la siguiente forma ``(start, stop, step)``
### Ejemplos
Se usa ``list`` para ver el range en forma de lista.

````python
#genera una lista con valores desde 0 a 5 incrementando de a uno en uno
list(range(6)) 
````

````python
#mismo ejemplo que el anterior pero especificando start en 0 e incrementando de a uno en uno
list(range(0,6,1)) 
````

````python
#generar una lista con los números desde 10 hasta 20 incrementando de a 1
list(range(10,21,1))
````

````python
#generar una lista con los números desde 10 hasta 20 saltando de dos en dos
list(range(10,21,2))
````

````python
#generar una lista con los números desde -7 hasta 7 incrementando de a 1
list(range(-7, 8)) 
````

````python
#generar una lista con los números desde 5 hasta 0 con decremento de 1
list(range(5,-8,-1))
````

## Colecciones ``range`` con ciclos ``for``

Los rangos se combinan perfectamente con la instrucción para (for):

Se define una variable que se utilizará para recorrer cada número en el rango. Si la variable es $i$, el ciclo ``for`` se puede leer **“Para cada número $i$ en el rango dado”.**

###Ejemplos

````python
# imprimir de forma creciente los números enteros desde −5 hasta 5

for i in range(-5, 6, 1):
  print(i)
````

````python
# imprimir de forma decreciente los números enteros desde 10 hasta 1

for i in range(10, 0, -1):
  print(i)
````

````python
#  imprimir de forma creciente los múltiplos de 5 desde 0 hasta 50

for i in range(5, 51, 5):
  print(i)
````

#### Ejemplo: La suma de los primeros $n$ números naturales

Se pueden crear dos funciones diferentes en python, que permiten calcular la suma de los primeros $n$ números naturales, es decir, permiten calcular la expresión

$$
1+2+3+ \cdots + (n-1) + n,\ \text{que abreviadamante se escribe como}\ \sum_{i=1}^n i
$$

Usando el ciclo ``while``
```
  def suma(n):
    s = 0
    i = 1
    while(i <= n):
      s += i
      i += 1
    return s 
```

Usando el ciclo ``for``
```
  def suma(n):
    s = 0
    for i in range(1, n + 1):
      s += i
    return s
```

La variable $i$ juega el rol de una **variable contadora** ya que permite llevar el conteo de cuantos ciclos se han efectuado.

La variable $s$ juega el rol de una **variable acumuladora** pues acumula o almacena el valor parcial que se desea calcular utilizando el ciclo.


````python
def suma(n):
  s = 0
  for i in range(1, n + 1):
    s += i
  return s

def main():
  n = int(input("n? = "))
  print("La suma de los primeros n números es:", end=" ")
  print(suma(n))

main()
  
````

## For anidados
Es un ciclo dentro de otro ciclo.

### Ejemplo
Se tienen dos listas una de adjetivos y otra de frutas, imprimir por cada adjetivo cada fruta

````python
adj = ["rojo", "grande", "hermoso"]
frutas = ["manzana", "pera", "guayaba"]

for x in adj:
  for y in frutas:
    print(x, y)
````
